```python
#标记班组
class3=['姚明君','周航','张瑜珊','白杨','何鑫','何治芹','魏宏亚','杨思雨','张青','陈璐','段永帅','薄珍珍','陈京京','冯喆','张志超','李露','梁钰凤','周奇','莫孟泽','周盼','张春苑','晁亚倩']
class4=['李健','邾冬梅','彭洁','王晴晴','李永萍','杨丽霞','洪进','向静','徐婵','陈飞燕','姚忠丽','王凯','杨蕊','孙启磊','于维纤','冯谷佳','杨梦情','王莉','贺嘉贝娜','陈金香','王丽平']
class5=['吉荣欣','王告先','姜文思','王盼盼','覃艳华','宫汝佳','赵伟','赵智强','孟朝君','李彦序','张波','丁伟君','李梦雅','张丽丽','孙强','彭晓珍','王莉飞','曹山']
df['is3']=[n in class3 for n in df['姓名']]
df['is4']=[n in class4 for n in df['姓名']]
df['is5']=[n in class5 for n in df['姓名']]
df['c']=0
#标记班组的另一种实现方法
df.loc[[n in class3 for n in df['姓名']],'c']=3
df.loc[[n in class4 for n in df['姓名']],'c']=4
df.loc[[n in class5 for n in df['姓名']],'c']=5
df.head()
```

```python
print(sum(df['is3']==True))
print(sum(df['is4']==True))
print(sum(df['is5']==True))
```

```python
#表示成字典没啥用
result={'3':{},'4':{},'5':{}}
grouped=df[df['c']!=0].groupby(by=['c','姓名'])
for i,g in grouped:
    result[str(i[0])][i[1]]=g.shape[0]
print(result)
```

```python
#还是DataFrame靠谱
r={'name':[],'class':[],'count':[]}
grouped=df[df['c']!=0].groupby(by=['c','姓名'])
for i,g in grouped:
    r['name'].append(i[1])
    r['class'].append(i[0])
    r['count'].append(g.shape[0])
d=pd.DataFrame(r)
d.sort_values(by='count',ascending=False)
```

```python
import numpy as np
print(d.groupby('class')['count'].agg([np.sum, np.mean, np.std]))
```

```python
#只看前三天试试
r={'name':[],'class':[],'count':[]}
df=df[df['时间']<'06/04/2017 00:00:00']
grouped=df[df['c']!=0].groupby(by=['c','姓名'])
for i,g in grouped:
    r['name'].append(i[1])
    r['class'].append(i[0])
    r['count'].append(g.shape[0])
d=pd.DataFrame(r)
d.sort_values(by='count',ascending=False)
```
